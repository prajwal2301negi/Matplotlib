```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Making data from 0-10
var1 =np.arange(0,10)
# Storing data in va1
```

```python
var2 = var1**2
print(var2)
```

```python
plt.plot(var1, color = 'red', label='Price of Gold')
plt.plot(var2, color = 'blue', label='Price of Prices')
# Plotting x^2 Graph
plt.legend(title='Difference in price')
plt.xlabel('Gold')
plt.ylabel('Price')
plt.title('Gold vs Prices')
```

```python
plt.plot(var1, color = 'red', label='Price of Gold', linestyle = '--')
plt.plot(var2, color = 'blue', label='Price of Prices', linestyle='-.')
# Plotting x^2 Graph
plt.legend(title='Difference in price')
plt.xlabel('Gold')
plt.ylabel('Price')
plt.title('Gold vs Prices')
```

```python
# PIECHART
```

```python
subjects = ('Maths', 'Phy', 'Chem', 'Eng', 'ED')
marks = [90, 92, 89, 86, 97]
```

```python
plt.pie(marks, labels=subjects)
```

```python
plt.pie(marks, labels=subjects)
plt.show()
# Will Remove the details below
```

```python
plt.pie(marks, labels=subjects, explode = (0.1, 0.2, 0.05, 0, 0.1))
plt.show()
# Explode will act like a bomb such that entity moves away
# In bracket are the percentage. Eg-> Physics move away by 20Percent, Chemistry by 5Percent
```

```python
plt.pie(marks, labels=subjects, explode = (0.1, 0.2, 0.05, 0, 0.1), shadow = True)
plt.show()
# Will create shadow
```

```python
plt.pie(marks, labels=subjects, explode = (0.1, 0.2, 0.05, 0, 0.1), shadow = True, radius=4)
plt.show()
# Will make our PieChart Big
```

```python
plt.pie(marks, labels=subjects, explode = (0.1, 0.2, 0.05, 0, 0.1), shadow = True, autopct='%.2f')
plt.show()
# Write marks in percentage form. Autopct represents number in percentage
#.2f-> we want decimal value upto 2 values
```

```python
dist1 = np.random.randn(10000)
dist2 = np.random.randn(10000)
dist2 = (dist2 * 10) + 50

# Histograms
plt.hist(dist1, bins=100, color='teal', alpha=0.8, label='dist1')
plt.hist(dist2, bins=100, color='blue', alpha=0.6, label='dist2')

# Legend
plt.legend()

# Plot
plt.show()
```
